```python
import pandas as pd
```

```python
pd.set_option('min_rows', 50)
```

```python
file = "../../data/2024_O1_I1_AUT.CSV"
column_names = ["chapa", "comprimento chapa (mm)", "largura chapa (mm)", "material chapa", 
                "espessura chapa (mm)", "quantidade chapa", "area chapa (mm2)", "peso chapa (kg)",
                "eficiencia chapa (%)", "quantidade golpes", ""]
columns_to_int = ["comprimento chapa (mm)", "largura chapa (mm)",
                  "quantidade chapa", "quantidade golpes"
                ]
```

```python
autonest_df = pd.read_csv(file, encoding='latin')
```

```python
print(autonest_df)
```

```python
first_column = autonest_df.iloc[:, 0]
```

```python
print(first_column)
```

```python
#first_column = first_column.astype(int, errors='ignore')
```

```python
#print(first_column)
```

```python
autonest_df.iloc[:, 0] = pd.to_numeric(first_column, errors='coerce')
```

```python
autonest_df.iloc[:, 0].dropna()
```

```python
index_rows = autonest_df.index[first_column == autonest_df.iloc[:, 0]].tolist()
```

```python
autonest_df = autonest_df.filter(items=index_rows, axis=0)
```

```python
autonest_df = autonest_df.loc[first_column > 0]
```

```python
autonest_df.columns = column_names
```

```python
autonest_df = autonest_df.drop("", axis=1)
```

```python
print(autonest_df[columns_to_int])
```

```python
autonest_df["comprimento chapa (mm)"] = pd.to_numeric(autonest_df["comprimento chapa (mm)"], errors='coerce')
print(autonest_df["comprimento chapa (mm)"])
```

```python
autonest_df[columns_to_int] = pd.to_numeric(columns_to_int, errors='coerce')
```

```python
print(autonest_df[columns_to_int])
```

```python
autonest_df = autonest_df.groupby(["comprimento chapa (mm)", "largura chapa (mm)", "espessura chapa (mm)", "material chapa"]).sum()
```
